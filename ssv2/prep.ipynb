# Create CSV for dataset subset

```python
import json 

with open('labels.json', 'r') as f:
    label_encoder = json.load(f)
```

```python
import os
import csv
def get_last_part_of_filename(file_path):
    filename = os.path.basename(file_path)
    filename_without_extension = os.path.splitext(filename)[0]
    return filename_without_extension.split('/')[-1]
def write_line_to_csv(file_path, line):
    with open(file_path, mode='a', newline='') as file:
        writer = csv.writer(file, delimiter=' ')
        writer.writerow(line)
```

```python
root_directory = 'mp4'

for root, directories, files in os.walk(root_directory):
    folder = get_last_part_of_filename(root)
    if folder != 'mp4':
        label = folder
        for file in files:
            line = ['ssv2/mp4/' + file,  label_encoder[label]]
            write_line_to_csv('ssv2_sub.csv', line)
```

```python
import shutil
for root, directories, files in os.walk(root_directory):
    folder = get_last_part_of_filename(root)
    if folder != 'mp4':
        for file in files:
            source = os.path.join(root, file)
            destination = ('mp4/')
            shutil.move(source, destination)
        shutil.rmtree(root)
```

```python
for root, dirs, files in os.walk('mp4'):
    print("Current directory:", root)
    for file in files:
        file_path = os.path.join(root, file)
        directory = os.path.dirname(file_path)
        print("Directory of", file, ":", directory)
```
